```python
import cv2

import pytesseract
import  numpy as np
import pyttsx3
import say
import os
from pytesseract import image_to_string
from gtts import gTTS

pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files (x86)\\Tesseract-OCR\\tesseract.exe'
```

```python
# The text that you want to convert to audio 
#mytext = 'Welcome to geeksforgeeks!ndlkvnosadnvoasfajsdfjafjsdfasdfsdmnoejfoejfpwepffjeojqjo'
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('rate', 130)    # Aquí puedes seleccionar la velocidad de la voz
engine.setProperty('voice',voices[0].id)
def recogonize(im):
    img=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    (thresh, im_bw) = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    im_bw=cv2.cvtColor(im_bw,cv2.COLOR_GRAY2BGR)
    mytext=image_to_string(im_bw)
# Language in which you want to convert 
    language = 'en'
    
    engine.say(mytext)
    engine.runAndWait()
    return mytext
```

```python
cap=cv2.VideoCapture(0)
skipframe=0
while True:
    ret,frame=cap.read()
    cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    if(skipframe%30==0):
        temp=recogonize(frame)
        print(temp)
    cv2.imshow('frame',frame)
    skipframe+=1
    if cv2.waitKey(1)&0xff==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
    
```
